
# Regression Models - From Linear to Ensemble Methods

This notebook covers:

1. What regression is
2. Linear and polynomial regression
3. Regularization (Ridge, Lasso, Elastic Net)
4. Tree-based regression models
5. Ensemble methods (Random Forest, Gradient Boosting)
6. Evaluation metrics

## What is Regression?

### Problem

We want to predict a **continuous value**:

    x → y (real number)

Examples:
- temperature prediction
- energy output
- reaction yield


### Difference vs Classification

Classification --> discrete labels  
Regression --> continuous values  


### Goal

Learn a function:

    f(x) ≈ y

```python

# DATA PREPARATION

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression

# Generate synthetic dataset
X, y = make_regression(
    n_samples=200,
    n_features=1,
    noise=20,
    random_state=42
)

# Convert to DataFrame
X = pd.DataFrame(X, columns=["feature"])
y = pd.Series(y)

# Visualize
plt.scatter(X, y)
plt.title("Regression Dataset")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```

## Linear Regression

### Idea

Fit a line:

    y = w·x + b

### Optimization

Minimize:

    Mean Squared Error (MSE)

### Strength

Simple and interpretable

### Weakness

Cannot model nonlinear relationships

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

preds = model.predict(X)

plt.scatter(X, y)
plt.plot(X, preds, color="red")
plt.title("Linear Regression")
plt.show()
```

## Polynomial Regression

### Idea

Transform input:

    x --> [x, x², x³, ...]

Apply linear regression on transformed features.


### Effect

Allows nonlinear curves.


### Risk

High-degree polynomials -->  overfitting

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

preds = model.predict(X_poly)

plt.scatter(X, y)
plt.scatter(X, preds, color="red")
plt.title("Polynomial Regression")
plt.show()
```

## Regularization


### Problem

Linear models can overfit when too flexible.


### Solution

Add penalty to weights:


### Ridge (L2)

Penalizes large weights

--> smooth model


### Lasso (L1)

Forces some weights to zero

--> feature selection


### Elastic Net

Combination of Ridge and Lasso


```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet

models = {
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.1),
    "ElasticNet": ElasticNet(alpha=0.1)
}

for name, model in models.items():
    model.fit(X, y)
    preds = model.predict(X)
    
    plt.scatter(X, y)
    plt.plot(X, preds)
    plt.title(name)
    plt.show()
```

## Decision Tree Regressor


### Idea

Split feature space into regions

Each region predicts a constant value.


### Strength

Captures nonlinear patterns easily

### Weakness

Overfits easily -->  unstable


```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=4)
model.fit(X, y)

preds = model.predict(X)

plt.scatter(X, y)
plt.scatter(X, preds, color="red")
plt.title("Decision Tree Regressor")
plt.show()
```

## Random Forest Regressor

### Idea

Combine many decision trees

### Mechanism

- each tree sees different data
- predictions are averaged

### Effect

- reduces variance
- more stable than single tree

```python
from sklearn.ensemble import RandomForestRegressor 
model.fit(X, y)

preds = model.predict(X)

plt.scatter(X, y)
plt.scatter(X, preds, color="red")
plt.title("Random Forest Regressor")
plt.show()
```

## Gradient Boosting Regressor

### Idea

Build models sequentially

Each new model:

    corrects previous errors

### Mechanism

- start with simple model
- fit residuals (errors)
- repeat

### Effect

- very powerful
- can model complex relationships

### Risk

can overfit if not controlled

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor()
model.fit(X, y)

preds = model.predict(X)

plt.scatter(X, y)
plt.scatter(X, preds, color="red")
plt.title("Gradient Boosting Regressor")
plt.show()
```

## Regression Metrics

### MSE (Mean Squared Error)

Penalizes large errors strongly

### MAE (Mean Absolute Error)

More robust to outliers

### R² Score

Explains variance:

- 1 --> perfect
- 0 --> baseline

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

preds = model.predict(X)

mse = mean_squared_error(y, preds)
mae = mean_absolute_error(y, preds)
r2 = r2_score(y, preds)

print("MSE:", mse)
print("MAE:", mae)
print("R²:", r2)
```
