## What is Dimensionality Reduction?

### Core Idea

Reduce number of features while preserving useful information.

### Why this matters

High-dimensional data:

- slow to process  
- noisy  
- harder to model  

### Goals

- faster training  
- less overfitting  
- remove noise  
- enable visualization  

### Curse of Dimensionality

As dimensions increase:

- distances lose meaning  
- data becomes sparse  
- models need more data  

### Key Insight

More features =/= more information

## Feature Selection vs Feature Extraction

### Feature Selection

Keep existing features:

- remove useless ones  
- choose important ones  

### Feature Extraction

Create new features:

- combine existing ones  
- transform data  

Example:
PCA creates new variables.

### Takeaway

Selection --> filtering  
Extraction --> transformation

## Feature Selection Methods

### Filter

Based on statistics (variance, correlation)

### Wrapper

Test model performance (RFE)

### Embedded

Built into models:

- Lasso --> removes features  
- Trees --> importance scores  

### Takeaway

Selection = remove irrelevant features

```python
#just getting the dataset here
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
data = load_iris()

# features + target
X = data.data
y = data.target
```

```python
# FILTER METHODS
from sklearn.feature_selection import VarianceThreshold

# remove low-variance features
selector = VarianceThreshold(threshold=0.1)

X_new = selector.fit_transform(X)

print("Original shape:", X.shape)
print("Reduced shape:", X_new.shape)
```

```python
# WRAPPER/ EMBEdDED METHODS
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)

print(X_new.shape)
```

## Why Scaling Matters

I repeat scaling this often, because its simply this important. IMO.

Dimensionality reduction depends on:

--> variance  
--> distance  

If features have different scales:

--> dominant features distort results  

### Key Rule

Always scale before PCA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Principal Component Analysis (PCA)

### Core Idea

Find directions with maximum variance.

### What PCA does

- rotates data  
- finds new axes  
- orders them by importance  

### Concepts

Principal Components:
    new directions  

Explained Variance:
    how much information retained  

Orthogonality:
    components are independent  

### Key Insight

PCA =

    rotate data to align with variance


```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
plt.title("PCA Projection")
plt.show()
```

## Explained Variance

Each component explains part of the total variance.

### Example

[0.7, 0.2]

--> first component = 70% information  
--> second = 20%  

### Goal

Choose components that capture most variance

```python
print(pca.explained_variance_ratio_)
```

## Kernel PCA

### Problem

PCA is linear.

--> cannot capture nonlinear patterns


### Solution

Transform data to higher space:

--> then apply PCA  

### Kernels

- RBF (most common)  
- polynomial  
- sigmoid  

### Takeaway

Kernel PCA =

    nonlinear version of PCA


```python
from sklearn.decomposition import KernelPCA

kpca = KernelPCA(n_components=2, kernel="rbf")
X_kpca = kpca.fit_transform(X_scaled)

plt.scatter(X_kpca[:,0], X_kpca[:,1], c=y)
plt.title("Kernel PCA")
plt.show()
```

## Singular Value Decomposition (SVD)

### Idea

Decompose matrix into components:

    X = U Σ Vᵀ

### Interpretation

- identifies latent factors  
- reduces dimensions  

### When to use

- sparse data (text, embeddings)

### Takeaway

SVD =

    matrix factorization for dimensionality reduction

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X_scaled)

plt.scatter(X_svd[:,0], X_svd[:,1], c=y)
plt.title("SVD")
plt.show()
```

## Visualisation
## t-SNE

### Idea

Preserve local relationships.

### Strength

Excellent visualization of clusters.

### Limitation

- slow  
- not suitable for preprocessing  

### Takeaway

t-SNE =

    visualization tool, not feature transformer

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y)
plt.title("t-SNE")
plt.show()
```

## Manifold Learning

### Idea

Data lies on low-dimensional structure.

### Methods

- Isomap --> geodesic distances  
- LLE --> local reconstruction  
- MDS --> distance preservation  

### Takeaway

Real-world data often has hidden structure

```python
from sklearn.manifold import Isomap

iso = Isomap(n_components=2)
X_iso = iso.fit_transform(X_scaled)
```

## Practical Workflows

### PCA before classification

### PCA before clustering

X --> scale --> PCA --> K-Means  

### Visualization

### Takeaway

Dimensionality reduction is part of pipeline

## Choosing the Right Method

| Goal | Method |
|------|-------|
| General reduction | PCA |
| Nonlinear patterns | Kernel PCA |
| Sparse data | SVD |
| Visualization | t-SNE |
| Large datasets | Incremental PCA |
