# Bayesian Optimization for Machine Learning Pipelines

Focus:

- Efficient hyperparameter optimization
- Replacing grid search with smarter search
- Practical usage with pipelines

Goal:

Understand how to search hyperparameters efficiently
for better model performance.


```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    # scale → stable optimization
    ("scaler", StandardScaler()),
    
    # PCA → reduce dimensions / noise
    ("pca", PCA()),
    
    # classifier
    ("clf", LogisticRegression())
])
```

## Bayesian Optimization - Core Idea

### Problem

Hyperparameter search space is large:

Grid Search:

    tries everything --> slow



### Bayesian Optimization

Instead of searching blindly:

--> learns from previous trials


### Process

1. Try some hyperparameters  
2. Measure performance  
3. Build a model of performance  
4. Predict promising regions  
5. Search there  


### Key Insight

Search becomes:

    adaptive and efficient

instead of:

    exhaustive

## Grid vs Bayesian Search Space

Grid Search:

    discrete values only

Example:
    C = [0.1, 1, 10]


Bayesian Optimization:

    continuous ranges

Example:
    C ∈ [0.001, 100]


### Advantage

Bayesian can search:

    infinite parameter space


```python
# BAYESIAN SEARCH SPACE

# requires:
# pip install scikit-optimize

from skopt import BayesSearchCV

search_space = {
    
    # continuous parameter (log scale!)
    # WHY: spans orders of magnitude
    "clf__C": (1e-3, 100, "log-uniform"),
    
    # discrete but treated as range
    "pca__n_components": (2, 4),
    
    # categorical choices
    "clf__solver": ["lbfgs", "saga", "newton-cg"],
    
    # optimization iterations
    "clf__max_iter": (100, 1000)
}

bayes = BayesSearchCV(
    pipeline,
    search_space,
    n_iter=20,
    cv=5,
    random_state=42,
    n_jobs=-1
)

bayes.fit(X, y)
```

```python
print("Best parameters:", bayes.best_params_)
print("Best score:", bayes.best_score_)
```

## What Just Happened?

For each iteration:

1. Model tried parameters  
2. Evaluated using cross-validation  
3. Learned which regions perform well  
4. Focused search on promising areas  


### Difference to Grid

Grid:
    static search

Bayesian:
    adaptive search


### Key Insight

Each iteration becomes smarter

# GRID SEARCH 
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(
    pipeline,
    {
        "pca__n_components": [2, 3, 4],
        "clf__C": [0.1, 1, 10],
        "clf__solver": ["lbfgs", "liblinear"]
    },
    cv=5
)

grid.fit(X, y)

print("Grid score:", grid.best_score_)
print("Bayesian score:", bayes.best_score_)

## Grid vs Bayesian

Grid Search:

- systematic
- expensive
- inefficient


Bayesian Optimization:

- learns from past results
- fewer evaluations
- better for large spaces


### Key Insight

Bayesian is better when:

- parameter space is large
- evaluations are expensive

## When NOT to Use Bayesian Optimization


- very small search space --> grid is enough  
- very cheap models --> brute force okay  
- debugging phase --> use grid or manual
